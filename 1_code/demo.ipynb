## Demo notebook for testing trained models. 
- This notebook shows how to train, test and evaluate deep learning models for the problem of Vehicle Classification from images. There are total of 196 classes, and about 8000 training images.

#### Import required packages

```python
import os
import cv2
import matplotlib.pyplot as plt

from testModel import TestModel
from data import get_imgpath_labels

from config import *
```

### Read model and prepare test class

```python
model_path = '../2_outputs/trainedModels/model.h5'
test_matfile = 'devkit/cars_test_annos.mat'

# We are only testing, not training. We don't have labels for testing set
img_names_list, _ = get_imgpath_labels(test_matfile, train=False)

testModel = TestModel(model_path)
```

### Take each image and predict its class

```python
img_idx = 2000 # int within 8000

img_path = os.path.join(DATA_DIR, IMG_DIR, img_names_list[img_idx])

pred_class, conf_class, class_name = testModel.get_preds(img_path)
```

### Display the image with class name

```python
plt.figure(figsize=(12,8))
plt.imshow(cv2.imread(img_path))
plt.title('Class: {0:s}  |  Confidence: {1:.02f}'.format(class_name, conf_class), fontsize=15)
plt.show()
```
